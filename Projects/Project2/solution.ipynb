```python
import pandas as pd
```

Read data from file

```python
df = pd.read_csv('adult.data.csv')
df.head()
```

How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.

```python
race_count = df['race'].value_counts()
race_count
```

What is the average age of men?

```python
average_age_men = df.loc[df['sex'] == 'Male', 'age'].mean().round(1)
average_age_men
```

What is the percentage of people who have a Bachelor's degree?

```python
percentage_bachelors = round((df.loc[df['education'] == 'Bachelors'].shape[0] / df.shape[0]) * 100, 1)
percentage_bachelors
```

What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?

What percentage of people without advanced education make more than 50K?

```python
higher_education = df.loc[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
higher_education.head()
```

```python
lower_education = df.loc[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
lower_education.head()
```

```python
higher_education_rich = round((higher_education.loc[higher_education['salary'] == '>50K'].shape[0] / higher_education.shape[0]) * 100, 1)
higher_education_rich
```

```python
lower_education_rich = round((lower_education[lower_education['salary'] == '>50K'].shape[0] / lower_education.shape[0]) * 100, 1)
lower_education_rich
```

What is the minimum number of hours a person works per week (hours-per-week feature)?

```python
df.head()
```

```python
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

What percentage of the people who work the minimum number of hours per week have a salary of >50K?

```python
num_min_workers = df[df['hours-per-week'] == min_work_hours]
num_min_workers.head()
```

```python
rich_percentage = int((num_min_workers[num_min_workers['salary'] == '>50K'].shape[0] / num_min_workers.shape[0]) * 100)
rich_percentage
```

What country has the highest percentage of people that earn >50K?

```python
people_per_country = df['native-country'].value_counts()
people_per_country_more50k = df.loc[(df['salary'] == '>50K'), 'native-country'].value_counts()
```

```python
highest_earning_country = (people_per_country_more50k / people_per_country).sort_values(ascending=False).index[0]
highest_earning_country
```

```python
highest_earning_country_percentage =  round((people_per_country_more50k / people_per_country).sort_values(ascending=False).values[0] * 100, 1)
highest_earning_country_percentage
```
